```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = sns.load_dataset("Iris")
```

```python
data.isnull().sum()
```

```python
data.dtypes
```

```python
for col in data.columns:
    if data[col].dtypes == 'object':
        print("Nominal :", col)
    else :
        print("Numeric :", col)
```

```python
data.hist()
plt.title("Histogram for each Features")
plt.show()
```

```python
sns.boxplot(data=data)
plt.title("Boxplot for all Features")
plt.show()
```

```python
sns.boxplot(x="sepal_length", data=data)
plt.title("SepalLength Boxplot")
plt.show()
```

```python
sns.boxplot(x="sepal_width", data=data)
plt.title("SepalWidth Boxplot")
plt.show()
```

```python
sns.boxplot(x="petal_length", data=data)
plt.title("PetalLength Boxplot")
plt.show()
```

```python
sns.boxplot(x="petal_width", data=data)
plt.title("PetalWidth Boxplot")
plt.show()
```
