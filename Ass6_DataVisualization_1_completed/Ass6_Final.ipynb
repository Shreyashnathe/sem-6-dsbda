```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset (inbuilt iris dataset)
df = sns.load_dataset('iris')

# -----------------------------
# 1. Features and their types
# -----------------------------
print("Features and their data types:\n")
print(df.dtypes)

print("\nCategorizing features:\n")

for col in df.columns:
    if df[col].dtype == 'object':
        print(col, "-> Nominal (Categorical)")
    else:
        print(col, "-> Numeric")
```

```python
# -----------------------------
# 2. Histogram for each feature
# -----------------------------
df.hist()
plt.suptitle("Histogram of all Features")
plt.show()
```

```python
# -----------------------------
# 3. Boxplot for each feature
# -----------------------------
sns.boxplot(data=df)
plt.title("Boxplot of all Features")
plt.show()
```

```python
# -----------------------------
# Boxplot per feature (better clarity)
# -----------------------------
for col in df.select_dtypes(include=['float64', 'int64']).columns:
    sns.boxplot(x=df[col])
    plt.title(f"Boxplot of {col}")
    plt.show()
```

```python

# -----------------------------
# 4. Basic Statistical Summary
# -----------------------------
print("\nStatistical Summary:\n")
print(df.describe())
```
