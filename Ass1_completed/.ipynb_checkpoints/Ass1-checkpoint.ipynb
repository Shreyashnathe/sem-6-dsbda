```python
mean_car = data['Car'].mean()
```

```python
data['Car'] = data['Car'].fillna(value = mean_car)
```

```python
data['Car'] = data['Car'].astype('int')
data['Address'] = data['Address'].astype('string')
data['Rooms'] = data['Rooms'].astype('int')
```
