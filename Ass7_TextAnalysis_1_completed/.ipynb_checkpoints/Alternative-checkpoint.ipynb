```python
import nltk 
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk import pos_tag
from sklearn.feature_extraction.text import TfidfVectorizer

nltk.download("all")
```

```python
file1 = open("file1.txt")

text1 = file1.read()
print(text1)
```

```python
token1 = word_tokenize(text1)
print(token1)
```

```python
pos_tag1 = pos_tag(token1)
print(pos_tag1)
```

```python
stop_words = set(stopwords.words('english'))
print(stop_words)

filtered_word1 = [word for word in token1 if word.lower() not in stop_words]

print("\n", filtered_word1)
```

```python
stemmer = PorterStemmer()

stemmedWord = [stemmer.stem(word) for word in filtered_word1]
print(stemmedWord)
```

```python
lemmatizer = WordNetLemmatizer()

lemmatizedWord = [lemmatizer.lemmatize(word) for word in filtered_word1]
print(lemmatizedWord)
```

```python
document = [text1]

vectorizer = TfidfVectorizer()

tfidf_matrix = vectorizer.fit_transform(document)

terms = vectorizer.get_feature_names_out()

values = tfidf_matrix.toarray()
```

```python
print("Terms :\n", terms)
```

```python
print("Values :\n", values)
```
