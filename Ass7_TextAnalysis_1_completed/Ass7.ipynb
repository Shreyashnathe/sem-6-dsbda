```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk import pos_tag
from sklearn.feature_extraction.text import TfidfVectorizer

nltk.download('all')
```

```python
file1 = open("file1.txt")
file2 = open("file2.txt")

text1 = file1.read()
text2 = file2.read()
```

```python
print("Text1 Tokenization")
token1 = word_tokenize(text1)
print(token1)

print("\nText2 Tokenization")
token2 = word_tokenize(text2)

print(token2)
```

```python
print("Text1 pos tag")
pos_tag1 = pos_tag(token1)
print(pos_tag1)

print("\nText2 pos tag")
pos_tag2 = pos_tag(token2)
print(pos_tag2)
```

```python
stop_words = set(stopwords.words('english'))

print("Text1 Filtered Words")
filtered_words1 = [word for word in token1 if word.lower() not in stop_words]
print(filtered_words1)

print("\nText2 Filtered Words")
filtered_words2 = [word for word in token2 if word.lower() not in stop_words]
print(filtered_words2)
```

```python
stemmer = PorterStemmer()

print("Text1 Stemming")
stemmed_words1 = [stemmer.stem(word) for word in filtered_words1]
print(stemmed_words1)

print("\nText2 Stemming")
stemmed_words2 = [stemmer.stem(word) for word in filtered_words2]
print(stemmed_words2)
```

```python
lemmatizer = WordNetLemmatizer()

print("Text1 Lemmatization")
lemmatized_words1 = [lemmatizer.lemmatize(word) for word in filtered_words1]
print(lemmatized_words1)

print("\nText2 Lemmatization")
lemmatized_words2 = [lemmatizer.lemmatize(word) for word in filtered_words2]
print(lemmatized_words2)
```

```python
documents = [text1, text2]
```

```python
vectorizer = TfidfVectorizer()

tfidf_matrix = vectorizer.fit_transform(documents)
```

```python
terms = vectorizer.get_feature_names_out()
```

```python
values = tfidf_matrix.toarray()
```
